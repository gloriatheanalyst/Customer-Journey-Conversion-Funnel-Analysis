### Conversion by Device Type

I examined how conversion rates vary across different device types.

- **Purpose:**  
  To understand which devices are more likely to lead to purchases and inform device-specific optimizations.  

- **Observations:**  
  - Desktop, Mobile, and Tablet users may convert at different rates.  
  - Higher conversion on one device could indicate better user experience or easier navigation.  
  - This analysis helps prioritize design improvements for devices with lower conversions.

```python
# Conversion Metrics by Device Type

session_conversion = df.groupby(['SessionID', 'DeviceType'])['Purchased'].max().reset_index()

# Calculate conversion metrics by device type
device_conversion = session_conversion.groupby('DeviceType').agg(
    Sessions=('SessionID', 'count'),
    Conversions=('Purchased', 'sum')
).reset_index()

# Calculate conversion rate
device_conversion['ConversionRate'] = (device_conversion['Conversions'] / device_conversion['Sessions'])

# Add a percentage column 
device_conversion['ConversionRate_Pct'] = device_conversion['ConversionRate'] * 100

print(device_conversion)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='DeviceType', y='ConversionRate', data=device_conversion,
            hue='DeviceType',
            palette=["#6f1d1b", "#bb9457", "#432818"], legend=False)
plt.title("Conversion Rate by Device Type", fontsize=14)
plt.ylabel("Conversion Rate", fontsize=12)
plt.xlabel("Device Type", fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Print formatted results
print("Conversion by device:")
for index, row in device_conversion.iterrows():
    print(f"{row['DeviceType']}: {row['ConversionRate_Pct']:.2f}%")

device_conversion.head()
```

### Conversion by Referral Source

I analyzed how users from different referral sources convert into purchasers.

- **Purpose:**  
  To evaluate which marketing channels or sources bring the most valuable traffic.  
  Conversion rate = (# of sessions with purchase) / (total sessions from that source).

- **Observations:**  
  - Some referral sources may have higher conversion rates despite lower traffic, indicating higher quality leads.  
  - Sources with low conversion but high traffic may need better targeting or landing page optimization.  


```python
# Conversion metrics by ReferralSource

session_df = df.groupby(['SessionID', 'ReferralSource'])['Purchased'].max().reset_index()

# Calculate conversion metrics by referral source
referral_table = session_df.groupby('ReferralSource').agg(
    Sessions=('SessionID', 'count'),
    Conversions=('Purchased', 'sum')
).reset_index()

# Calculate conversion rate
referral_table['ConversionRate'] = (referral_table['Conversions'] / referral_table['Sessions']).round(4)

# Sort and Plot
sorted_data = referral_table.sort_values('ConversionRate', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='ReferralSource', y='ConversionRate', data=sorted_data,
            hue='ReferralSource',
            palette=["#6f1d1b","#bb9457","#432818","#99582a"], legend=False)
plt.title("Conversion Rate by Referral Source", fontsize=14)
plt.ylabel("Conversion Rate", fontsize=12)
plt.xlabel("Referral Source", fontsize=12)
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()

# Print formatted results
print("Conversion by referral source:")
for index, row in sorted_data.iterrows():
    print(f"{row['ReferralSource']}: {row['ConversionRate']*100:.2f}% ({int(row['Conversions'])}/{int(row['Sessions'])} sessions)")

# Display of the table with formatted percentages
display_df = referral_table.copy()
display_df['ConversionRate'] = display_df['ConversionRate'].apply(lambda x: f"{x*100:.2f}%")
display(display_df)
```

### Conversion and Cart Metrics by Country

I analyzed user behavior and conversions across different countries.

- **Purpose:**  
  To understand regional differences in browsing, cart activity, and purchases.  
  Metrics include:
  - **Sessions:** Total sessions from the country.  
  - **Cart Rate:** Proportion of sessions where at least one item was added to the cart.  
  - **Conversion Rate:** Proportion of sessions ending in purchase.  
  - **Cart To Conversion Rate:** Proportion of cart sessions that converted to purchase.  
  - **Abandonment Rate:** Proportion of cart sessions that did not convert.

- **Observations:**  
  - Some countries may have high traffic but low conversion, suggesting potential UX or payment issues.  
  - Cart-to-conversion and abandonment rates reveal friction points in the checkout process.  


```python
# Sessions and Conversions by country

sessions_by_country = df.groupby('Country')['SessionID'].nunique().sort_values(ascending=False)

# Plot sessions by country

plt.figure(figsize=(12, 6))
sns.barplot(x=sessions_by_country.index, y=sessions_by_country.values,
            hue=sessions_by_country.index,
            palette=sns.color_palette(["#ffedd8","#f3d5b5","#e7bc91","#d4a276","#bc8a5f","#a47148","#8b5e34","#6f4518","#603808","#583101"], len(sessions_by_country)), legend=False)
plt.title("Number of Sessions by Country", fontsize=14)
plt.ylabel("Number of Sessions", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()

# Print sessions by country

print("Sessions by Country:")
for country, count in sessions_by_country.items():
    print(f"{country}: {count:,}")

# Aggregating key metrics by Sessions and Country
session_country_df = df.groupby(['SessionID', 'Country']).agg(
    ItemsInCart=('ItemsInCart', 'max'),
    Purchased=('Purchased', 'max')
).reset_index()

# Create a table aggregating by Country
country_table = session_country_df.groupby('Country').agg(
    Sessions=('SessionID', 'count'),
    CartSessions=('ItemsInCart', lambda x: (x > 0).sum()),  # Sessions with items in cart
    Conversions=('Purchased', 'sum')
).reset_index()

# Calculate additional metrics
country_table['CartRate'] = (country_table['CartSessions'] / country_table['Sessions']).round(4)
country_table['ConversionRate'] = (country_table['Conversions'] / country_table['Sessions']).round(4)
country_table['CartToConversionRate'] = (country_table['Conversions'] / country_table['CartSessions']).fillna(0).round(4)
country_table['AbandonmentRate'] = ((country_table['CartSessions'] - country_table['Conversions']) / country_table['CartSessions']).fillna(0).round(4)

# Sort by conversion rate in descending order
country_table = country_table.sort_values('ConversionRate', ascending=False)


# Plot for conversion rates
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='ConversionRate', data=country_table,
            hue='Country',
            palette=["#ffedd8","#f3d5b5","#e7bc91","#d4a276","#bc8a5f","#a47148","#8b5e34"], legend=False)
plt.title("Conversion Rate by Country", fontsize=14)
plt.ylabel("Conversion Rate", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()

# Create a multi-metric comparison

colors = ["#6f1d1b","#bb9457","#432818","#99582a"]
plt.figure(figsize=(14, 8))
metrics = ['CartRate', 'ConversionRate', 'CartToConversionRate', 'AbandonmentRate']
melted_data = pd.melt(country_table, id_vars=['Country'], value_vars=metrics, 
                      var_name='Metric', value_name='Rate')

sns.barplot(x='Country', y='Rate', hue='Metric', data=melted_data, palette=colors)
plt.title("E-commerce Metrics by Country", fontsize=14)
plt.ylabel("Rate", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.xticks(rotation=0, ha='right')
plt.legend(title='Metric', loc='upper left', bbox_to_anchor=(1, 1), fontsize=8, title_fontsize=9)
plt.tight_layout()
plt.show()

# Display table
display_df = country_table.copy()
for col in ['CartRate', 'ConversionRate', 'CartToConversionRate', 'AbandonmentRate']:
    display_df[col] = display_df[col].apply(lambda x: f"{x*100:.2f}%")
display(display_df)
```

### Conversion Funnel Analysis

I analyzed the number of unique sessions at each page type to understand how users progress through the e-commerce funnel:  

- **Funnel Stages:** `home â†’ product_page â†’ cart â†’ checkout â†’ confirmation`  
- **Purpose:**  
  - Track user progression and drop-offs at each stage.  
  - Identify stages with high abandonment to target improvements.  

- **Metrics Calculated:**  
  - **Sessions:** Number of unique sessions at each stage.  
  - **Percentage:** Sessions at each stage as a % of total sessions starting at home.  
  - **DropOff:** Number of sessions lost compared to the previous stage.  
  - **DropOffRate:** Percentage of sessions lost from the previous stage.  

- **Observations:**  
  - The funnel visualizations clearly show where the largest drop-offs occur.  
  - Most sessions drop off between `cart â†’ checkout` and `checkout â†’ confirmation`.  
  - Understanding these drop-offs helps optimize the checkout flow and reduce friction.  


```python
# Count unique sessions at each PageType

funnel_counts = df.groupby('PageType')['SessionID'].nunique().reindex(
    ['home', 'product_page', 'cart', 'checkout', 'confirmation'])

# Dataframe created
funnel_df = pd.DataFrame({
    'Stage': funnel_counts.index,
    'Sessions': funnel_counts.values
})

# Calculate metrics
funnel_df['Percentage'] = (funnel_df['Sessions'] / funnel_df['Sessions'].iloc[0] * 100).round(2)
funnel_df['Previous'] = funnel_df['Sessions'].shift(1)
funnel_df['DropOff'] = funnel_df['Previous'] - funnel_df['Sessions']
funnel_df['DropOffRate'] = (funnel_df['DropOff'] / funnel_df['Previous'] * 100).round(2)

# Print funnel metrics
print("Conversion Funnel Analysis:")
print("--------------------------")
print(f"Total sessions starting at home page: {funnel_df['Sessions'].iloc[0]:,}")
print("\nFunnel Stages:")
for i, row in funnel_df.iterrows():
    if i == 0:
        print(f"{row['Stage']}: {row['Sessions']:,} sessions (100%)")
    else:
        prev_stage = funnel_df['Stage'].iloc[i-1]
        print(f"{row['Stage']}: {row['Sessions']:,} sessions ({row['Percentage']}% of total, " 
              f"{100-row['DropOffRate']:.2f}% from {prev_stage})")

# Bar plot with percentage labels
plt.figure(figsize=(10, 8))
ax = sns.barplot(x='Stage', y='Sessions', data=funnel_df,
                 hue='Stage',
                palette=["#6f1d1b","#bb9457","#432818","#99582a","#ffe6a7"], legend=False)
for i, row in funnel_df.iterrows():                       # Add percentage labels on top of bars
    ax.text(i, row['Sessions'] + (funnel_df['Sessions'].max() * 0.02), 
            f"{row['Percentage']}%", ha='center', fontweight='bold')
plt.title("Conversion Funnel: Sessions at Each Stage", fontsize=14)
plt.ylabel("Number of Sessions", fontsize=12)
plt.xlabel("Funnel Stage", fontsize=12)
plt.tight_layout()
plt.show()

# Plot funnel chart
plt.figure(figsize=(10, 8))

stages = funnel_df['Stage']
values = funnel_df['Sessions']
max_width = 0.8
widths = np.linspace(max_width, 0.4, len(stages))  # Decreasing widths

for i, (stage, value, width) in enumerate(zip(stages, values, widths)):
        left = (max_width - width)/2
        plt.bar(x=i, height=value, width=width, bottom=0, color=["#6f1d1b","#bb9457","#432818","#99582a","#ffe6a7"][i]) 
        plt.text(i, value + (values.max() * 0.02), f"{value:,}", ha='center', fontweight='bold')
        plt.text(i, value/2, f"{funnel_df['Percentage'].iloc[i]}%", ha='center', 
             va='center', color='white', fontweight='bold')

plt.xticks(range(len(stages)), stages)
plt.title("E-commerce Conversion Funnel", fontsize=14)
plt.ylabel("Number of Sessions", fontsize=12)
plt.xlabel("Funnel Stage", fontsize=12)
plt.tight_layout()
plt.show()

# Metrics table
display_df = funnel_df.copy()
display_df['Percentage'] = display_df['Percentage'].apply(lambda x: f"{x:.2f}%")
display_df['DropOffRate'] = display_df['DropOffRate'].apply(lambda x: f"{x:.2f}%" if not pd.isna(x) else "N/A")
display_df = display_df.fillna("N/A")
display(display_df)
```
